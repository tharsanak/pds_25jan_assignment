## Q1.1: Reading in rows of a CSV file. [10 marks]

Write a function `csv_rows(filename, from_row=None, to_row=None)`

The input arguments of the function should be a string specifying a filename,

The row numbers should be either integers or None.

You can use the file `part1.csv` for your tests.

### 1.1 - csv_rows() Function
This function reads a CSV file and returns the header and the rows I need. The first line of the file is the header, so I read that separately. After that, I go through the rest of the file one row at a time. Each value is split by commas, and if it can be converted into a number, I change it using `float()`. If it cannot be converted, like `-` for missing data, I leave it as text.

The `from_row` and `to_row` arguments let me choose which rows to return. The function uses 1‑based indexing, so row 1 refers to the first data row after the header. If `from_row` is not given, it defaults to 1. If `to_row` is not given, the function reads all rows until the end of the file.

For example, calling `csv_rows('part1.csv', from_row=2, to_row=4)` returns rows 2, 3, and 4 along with the header.



```python

def csv_rows(filename, from_row=None, to_row=None):
    
    # If from_row isn't given, start from row 1 
    if from_row is None:
        from_row = 1

    header = []
    rows = []

    with open(filename, 'r') as f:
        # Read the header line and split it into column names
        header_line = f.readline()
        header = header_line.strip().split(',')

        # Read all remaining rows from the file
        all_rows = []
        for row_text in f:
            row_text = row_text.strip()
            if not row_text:
                continue  # skip empty 

            raw_values = row_text.split(',')
            cleaned_row = []

            # Convert numbers where possible, otherwise keep as string
            for value in raw_values:
                value = value.strip()
                try:
                    cleaned_row.append(float(value))
                except ValueError:
                    cleaned_row.append(value)

            all_rows.append(cleaned_row)

    # Convert row numbers to list slice positions
    start = from_row - 1
    if to_row is None:
         end = len(all_rows)
    else:
         end = to_row

    rows = all_rows[start:end]

    return header, rows
```

Write a function `csv_cols(filename, from_row=None, to_row=None, columns=None)`
that takes as inputs a filename, a starting row, a final row and a list of column names.
The same conventions as for Q1.1 should be used for the starting and final row input parameters.
The column name parameter can either be a list of column names or None, which indicates all columns.

### 1.2 – csv_cols() Function

This function builds on csv_rows from Q1.1 to reorganise the data from rows into columns. I return a dictionary where each key is a column name and the value is a list of its corresponding entries. This makes the data easier to work with, as each variable can be accessed directly by name.

I first use csv_rows to get the header and the filtered rows. If no specific columns are requested, I just use all the columns in the file. If column names are provided, I check they exist in the header and raise an error if any are missing. For each selected column, I find its index in the header and pull out the values from each row.

For example, passing columns= ['10-14 years old','70+ years old'] returns only those two columns in the dictionary and ignores the remaining columns.

```python

def csv_cols(filename, from_row=None, to_row=None, columns=None):
   
    # Get the header and the rows from the file
    header, rows = csv_rows(filename, from_row=from_row, to_row=to_row)

    # If no columns are given, use all columns
    if columns is None:
        selected_columns = header
    else:
        # Check that each chosen column exists
        for column in columns:
            if column not in header:
                raise ValueError(f"Column '{column}' not found in the file.")
        selected_columns = columns

    # Create the dictionary using the values from each row
    result = {}
    for column in selected_columns:
        col_index = header.index(column)
        result[column] = [row[col_index] for row in rows]

    return result
```

```python

# Test 1: Read all columns for the first 3 rows
data = csv_cols('part1.csv', from_row=1, to_row=3)
print('Test 1 - all columns, rows 1 to 3:-')
for column, values in data.items():
    print(f'  {column}: {values}')
print()

# Test 2: Only read these two columns
data2 = csv_cols('part1.csv', from_row=1, to_row=3,
                 columns=['10-14 years old', '70+ years old'])
print('Test 2 - two selected columns:-')
for column, values in data2.items():
    print(f'  {column}: {values}')
```

## Q1.3: Calculating the minimum, maximum and median of a list of values [10 marks]

Write a function `min_max_median(p)` to calculate
the minimum, maximum and median of a list of values.
The function should return the minimum, maximum and median.




### 1.3 – min_max_median() Function

This function calculates the minimum, maximum, and median of a list of values. Before doing any calculations, I remove non numeric entries such as -, because these represent missing data and should not affect the results.

After filtering, I use Python’s built‑in min() and max() functions to find the smallest and largest values. To compute the median, I sort the numeric values first. If the list has an odd number of elements, the median is simply the middle value. If it has an even number of elements, I take the average of the two middle values.

If the list contains no numeric values at all, the function raises a ValueError instead of returning an incorrect result. This ensures the function only works with valid data.

```python
def min_max_median(values):
   
    # Keep only the numbers and ignore anything else
    numbers = [i for i in values if isinstance(i, (int, float))]

    if len(numbers) == 0:
        raise ValueError('No numeric values found in the list.')

    minimum = min(numbers)
    maximum = max(numbers)

    # Sort the numbers so we can find the middle value
    sorted_numbers = sorted(numbers)
    count = len(sorted_numbers)

    # If the list length is odd, take the middle number
    if count % 2 == 1:
        median = sorted_numbers[count // 2]
    # If even, average the two middle numbers
    else:
        median = (sorted_numbers[count // 2 - 1] + sorted_numbers[count // 2]) / 2

    return minimum, maximum, median
```

```python

# Test 1: Odd list — the middle value is the median
result = min_max_median([1.0, 3.0, 5.0, 7.0, 9.0])
print('Test 1 - odd length list:')
print(f'  min={result[0]}, max={result[1]}, median={result[2]}')
print()

# Test 2: Even list — median comes from averaging the two middle values
result2 = min_max_median([1.0, 2.0, 3.0, 4.0])
print('Test 2 - even length list:')
print(f'  min={result2[0]}, max={result2[1]}, median={result2[2]}')
print()

# Test 3: List contains a dash, which should be ignored
result3 = min_max_median([1.0, '-', 3.0, 4.0])
print('Test 3 - list with missing value:')
print(f'  min={result3[0]}, max={result3[1]}, median={result3[2]}')
```

## Q1.4: Calculating the Minimum, Maximum and Median of Rows and Columns of a Table [10 marks]

Write a function `table_min_max_median(data, columns=None)` that takes as inputs
a dictionary with column names as keys and a list of values for each key.
The `columns` parameter can be either a list of column names or None, which
means all column names should be used.
The function should return a list of (minimum, maximum, median) tuples for each row,
and a list of (minimum, maximum, median) tuples for the selected columns.
Your function must use the function from Q1.3.

You can use the file `part1.csv` and the function from Q1.2 for your tests.


### 1.4 – table_min_max_median() Function

This function takes the column‑based dictionary produced by `csv_cols` and calculates the minimum, maximum, and median in two directions: across each row and down each selected column. It uses my `min_max_median` function from Q1.3 to compute all the statistics.

I first decide which columns to include. If no column list is provided, all columns in the dictionary are used. If specific column names are given, I check that each one exists before continuing.

For the row statistics, I take one value from each selected column at the same row index and pass that list to `min_max_median`. This gives a (min, max, median) summary for each row, showing how the values vary across columns at that position.

For the column statistics, I take the full list of values from each selected column and send it directly to `min_max_median`. This shows how a single variable changes down the table.

The function returns two lists of tuples, one list for the row statistics and other list for the column statistics. Each tuple contains the minimum, maximum, and median. The `columns` parameter is helpful when I only want to analyse a subset of the table instead of all columns.


```python

def table_min_max_median(data, columns=None):
    
    # decide which columns to use
    if columns is None:
        selected_columns = list(data.keys())
    else:
        for column in columns:
            if column not in data:
                raise ValueError(f"Column '{column}' not found in data.")
        selected_columns = columns

    num_rows = len(data[selected_columns[0]])

    # calculate the stats for each row
    row_stats = []
    for i in range(num_rows):
        row_values = [data[column][i] for column in selected_columns]
        row_stats.append(min_max_median(row_values))

    # calculate the stats for each column
    column_stats = []
    for column in selected_columns:
        column_stats.append(min_max_median(data[column]))

    return row_stats, column_stats
```

```python

# Test 1: Example to check
print('Test 1 - example to check:-')
example = {
    'A': [1.0, 4.0, 7.0],
    'B': [2.0, 5.0, 8.0],
    'C': [3.0, 6.0, 9.0],
}
row_s, col_s = table_min_max_median(example)
print('Row stats:')
for i, s in enumerate(row_s):
    print(f'  Row {i+1}: min={s[0]}, max={s[1]}, median={s[2]}')
print('Column stats:')
for col, s in zip(example.keys(), col_s):
    print(f'  {col}: min={s[0]}, max={s[1]}, median={s[2]}')
print()

# Test 2: First 5 rows from the CSV
print('Test 2 - first 5 rows from the CSV:-')
data = csv_cols('part1.csv', from_row=1, to_row=5)
row_stats, col_stats = table_min_max_median(data)
print('Row stats:')
for i, s in enumerate(row_stats):
    print(f'  Row {i+1}: min={s[0]:.4f}, max={s[1]:.4f}, median={s[2]:.4f}')
print('Column stats:')
for col, s in zip(data.keys(), col_stats):
    print(f'  {col}: min={s[0]:.4f}, max={s[1]:.4f}, median={s[2]:.4f}')
```

Write a function `print_min_max_median(row_stats, col_stats, ...)`

### 1.5 – print_min_max_median() Function
This function takes the row and column statistics from Q1.4 and prints them in a clear, readable table. Each entry in row_stats and col_stats is a tuple containing the minimum, maximum, and median for that row or column.

I added a few optional parameters to make the output more flexible. The row_labels and col_labels arguments let me use meaningful names instead of generic labels like “Row 1”. The title parameter adds a heading at the top, and decimal_places controls how many decimal places the numbers are shown with.

To keep the table neat, I calculate the column widths based on the longest label so everything lines up properly. I use ljust() and rjust() to pad the text so the columns stay tidy and easy to read.

At the bottom of each section, I also print an OVERALL row. This shows the smallest minimum, the largest maximum, and the median of all the medians. It gives a quick summary of the whole dataset without needing to scan through every row.


```python
def print_min_max_median(row_stats, col_stats,
                         row_labels=None, col_labels=None,
                         title='Statistics Summary',
                         decimal_places=4):

    # Format the number
    def fmt(value):
        return ('{:.' + str(decimal_places) + 'f}').format(value)

    # Use default labels 
    if row_labels is None:
        row_labels = [f'Row {i + 1}' for i in range(len(row_stats))]
    if col_labels is None:
        col_labels = [f'Col {i + 1}' for i in range(len(col_stats))]

    # Calculate label widths
    row_label_width = max(len(label) for label in row_labels + ['Label'])
    col_label_width = max(len(label) for label in col_labels + ['Label'])
    stat_width = max(decimal_places + 4, 8)

    # Print title
    print()
    print('=' * 50)
    print(title.center(50))
    print('=' * 50)

    # Print section
    def print_section(heading, labels, stats, label_width):
        print()
        print(heading)
        header_row = (
            'Label'.ljust(label_width) + '  ' +
            'Min'.rjust(stat_width) + '  ' +
            'Max'.rjust(stat_width) + '  ' +
            'Median'.rjust(stat_width)
        )
        print(header_row)
        print('-' * len(header_row))

        for label, (mn, mx, med) in zip(labels, stats):
            print(
                label.ljust(label_width) + '  ' +
                fmt(mn).rjust(stat_width) + '  ' +
                fmt(mx).rjust(stat_width) + '  ' +
                fmt(med).rjust(stat_width)
            )

        print('-' * len(header_row))

        # overall summary 
        all_mins = [s[0] for s in stats]
        all_maxs = [s[1] for s in stats]
        all_meds = sorted([s[2] for s in stats])
        n = len(all_meds)
        overall_med = (
            (all_meds[n // 2 - 1] + all_meds[n // 2]) / 2
            if n % 2 == 0 else all_meds[n // 2]
        )
        print(
            'OVERALL'.ljust(label_width) + '  ' +
            fmt(min(all_mins)).rjust(stat_width) + '  ' +
            fmt(max(all_maxs)).rjust(stat_width) + '  ' +
            fmt(overall_med).rjust(stat_width)
        )

    print_section('ROW STATISTICS', row_labels, row_stats, row_label_width)
    print_section('COLUMN STATISTICS', col_labels, col_stats, col_label_width)
    print()
```

```python
# Test 1: First 5 rows
data = csv_cols('part1.csv', from_row=1, to_row=5)
col_names = list(data.keys())
row_stats, col_stats = table_min_max_median(data)

print_min_max_median(
    row_stats, col_stats,
    row_labels=['Row 1', 'Row 2', 'Row 3', 'Row 4', 'Row 5'],
    col_labels=col_names,
    title='part1.csv - Rows 1 to 5',
    decimal_places=4
)

# Test 2: Full dataset summary
data_all = csv_cols('part1.csv')
col_names_all = list(data_all.keys())
rs_all, cs_all = table_min_max_median(data_all)

print_min_max_median(
    rs_all, cs_all,
    col_labels=col_names_all,
    title='part1.csv - Full Dataset Summary',
    decimal_places=4
)
```
